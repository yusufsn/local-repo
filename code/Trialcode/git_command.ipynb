```python
import os, glob, csv, subprocess, sys, re
from git import *
from subprocess import Popen, PIPE
from os import path
import pandas as pd
```

```python
userhome = os.path.expanduser('~')
repo = userhome + r'/historage-mining/historage/hbase/'
```

```python
def exec_command(cmd,work_dir):
    pipe = subprocess.Popen(cmd, shell=True, cwd=work_dir, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    (out, error) = pipe.communicate()
    return out, error
    pipe.wait()
```

```python
git_cmd = "git log --date=short --pretty=format:%cd,%h --no-merges"
logs = (str(exec_command(git_cmd,repo)).replace("b'","").replace('b"',"").replace('(',"",1).replace("\\'","'").
       split("\\n"))

commit_logs = []
for item in logs:
    temp = []
    for element in item.split(','):
        temp.append(element)
    commit_logs.append(temp)
commit_logs
```
