```python
import json, csv
import pandas as pd
```

```python
list1=[]
with open("complete_ref.json") as f1:
    iffi= json.load(f1)
    for p in iffi:
        if p["references"]["github"]["pull"]!=[]:
            list1.append(p["references"]["github"]["pull"][0])
list1      
```

```python
with open("pull.csv","w") as myfile:
    adi = csv.writer(myfile)
    adi.writerow(['pullrequest'])
    for item in list1:
        adi.writerow([item])
```

```python
list2 = []
for link in list1:
    head, sep, tail = link.partition("/pull")
    temp = head[:8] + "api." + head[8:19] + "repos/" + head[19:]
    list2.append(temp)
list2
```

```python
df = pd.read_csv("pull.csv")
df
```

```python
df['repos'] = list2
df.to_csv("pull.csv", index=False)
```
